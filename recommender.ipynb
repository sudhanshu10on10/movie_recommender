#Data Preprocessing

```python
#columns to be used
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#Nulls
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
#Duplicates
print(movies.duplicated().sum())
```

```python
# Take only important info in the columns
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval
```

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):                       #converts the object to a list of strings
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Get the top actors in the movie
def convert3(obj):
  l=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!= 3:                       #converts the object to a list of strings
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(3)
```

```python
#Get the director of the movie
movies['crew'][0]
```

```python
def fetch_director(obj):
  l=[]
  for i in ast.literal_eval(obj):                       #converts the object to a list of strings
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(3)
```

```python
#covert overview to a list so that it's easy to merge with other columns
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['keywords'].info()
```

```python
# Remove spaces - kyuki tags alag nahi rakhna chaiye - SamWorthington than Sam Worthington
movies['genres']=movies['genres'].apply(lambda x : [i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x : [i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x : [i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x : [i.replace(' ','') for i in x])
```

```python
movies.head(3)
```

```python
#convert to a new df
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# list to string for tags
new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df['tags'][0]
```

```python
# convert all tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df
```

#Text Vectorization

##Convert tags to vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out()
```

##Stemming

```python
#Stemming - converts dancing, danced, etc to Dance
import nltk
```

```python
new_df['tags'][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
ps.stem('lovinging')
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

##Calculate cosine distance of vectors

```python
#euclidean distance is not a good measure in high dimensional data - cosine distance works here
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector).shape
```

```python
#Distance of every movie with every other movie
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
  return
```

```python
recommend('Spectre')
```
